```python
# PowerMax_School
# Created 09/08/23 - Uses data from polygon.io to get real-time option data foe trading and pricing.
#                  - Uses polygon.io historical data to build trade_pairs records for ML v2.0
# Scratchpad to learn about polygon JSON and python APIs.
#         09/11/23 - Pair Trade Builder works. Yowza!
#         09/10/23 - Little price grabber works. TA works. So lucky. Uninstalled Visual Studio to remove version fights.
#

# *****************
# Import Libraries
# *****************

import csv
import copy
import json
import ivolatility as ivol
import numpy as np
import os
import pandas as pd
import requests   # for http requests
import scipy
from   scipy import stats
from   scipy.stats import norm
import time

import polygon
from polygon import RESTClient
#from   polygon import (RESTClient, StocksClient, OptionsClient, build_option_symbol)
import yfinance as yf
import yahooquery as yq

import datetime
from   datetime import datetime
from   datetime import date
from   datetime import timedelta
#from   datetime import fromtimestamp

# **********
#
# Build trade_pairs df. 
# 1. Choose jpoin key. Use Option ID.
# 2. Select all first_time option records. Create col "Option Sequence" with 3 categories: first time, ongoing, and last day. first_time.
#      No need to update option records in real-time, since option ID will sort it out. 
# 3. Build big df with all option records. Delete all 'first time' options
# 4. Join first time option df with all_options df. This creates the trade_pair df.
# 5. Add 'Close Action' col. Trade Close is common. For expiry date close action is Expiry Close or Expiry Assignment.
#       For options > X% ITM, Close Action is 'Eff Assignment.' For > Y% OTM, 'Close Action/ is Eff Expiry.
# **********

# Append features to all_trades
#all_trades_feat = pd.merge(all_trades_test, features_test, on multi-index)

all_trades = pd.read_csv('all_trades_test_single.csv')
# Create test option_ticker
all_trades['option ticker'] = all_trades['root'].astype(str) + '$' + all_trades['strike'].astype(str) + \
                              all_trades['option type'].astype(str) + all_trades['expiry day'].astype(str)

sort_cols = ['option ticker', 'day num']
all_trades.sort_values(by = sort_cols, axis = 0, inplace = True, ignore_index = True)

all_trades['option day seq'] = all_trades.groupby(by=['option ticker'], sort = False).cumcount()
max_cum_count = all_trades['option day seq'].max()

# Prep reduced_all_trades to be the daily trade
drop_cols = ['root', 'expiry day', 'strike', 'option type', 'close action']
reduced_all_trades = all_trades.drop(labels = drop_cols, axis = 1).copy()

# Prep all_trades to become first trades#
rename_cols = {'option day seq': 'first option day seq', 'day num': 'first day num'}
all_trades.rename(columns = rename_cols, inplace = True)
#print(all_trades.columns)
#print(all_trades)
all_trades.to_csv('all_trades_cols_renamed.csv')

# Create all pair trade records with repeated merges using successively newer trade records as the first trade.
# The unique option tickers keep all trade records grouped for the merge and the option sequence number keeps
# the rwos in sequence.
pair_trade_dfs = {}
for i in range(max_cum_count):
    first_trades = all_trades[all_trades['first option day seq'] == i]
    reduced_all_trades = reduced_all_trades[reduced_all_trades['option day seq'] != i].copy()
    pair_trade = pd.merge(left=first_trades, right=reduced_all_trades, how = 'left', on = 'option ticker', copy = True)
    pair_trade_id = 'pair_trade' + str(i)
    #print('pair_trade_id = ', pair_trade_id)
    #print(pair_trade)
    pair_trade_dfs[pair_trade_id] = pair_trade.copy()
    #print('pair_trade_dfs = ')
    #print(pair_trade_dfs)
# TODO - Ask Ryan how to name the pair_trade files so the multi-merge will work. and why dict works?
pair_trade_values = pair_trade_dfs.values()
#print('pair_trade_values = ')
#print(pair_trade_values)

all_pair_trades = pd.concat(pair_trade_values, axis = 0, join = 'outer', keys = None, ignore_index = True)
# Remove all ZDTE trade pairs.
all_pair_trades = all_pair_trades[all_pair_trades['first day num'] != all_pair_trades['expiry day']].copy()

sort_cols = ['option ticker', 'first option day seq', 'option day seq']
all_pair_trades.sort_values(by = sort_cols, axis = 0, inplace = True, ignore_index = True)
all_pair_trades.to_csv('all_pair_trades_test.csv')

print('thats all, folks.')
```
